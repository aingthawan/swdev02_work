```python
from lxml import html

def extract_text(html_string):
    """Extracts text content from specific HTML tags using lxml parser"""
    # Define the tags to extract text content from
    tags_to_extract = ['p', 'h1', 'h2', 'h3', 'h4', 'h5', 'h6', 'a', 'ul', 'ol', 'li', 'div', 'span', 'strong', 'em', 'blockquote', 'cite', 'pre', 'br']
    # Create an empty list to store extracted text
    extracted_text = []
    # Parse the HTML string using the lxml parser
    doc = html.fromstring(html_string)
    # Loop through each tag to extract text content
    for tag in tags_to_extract:
        # Find all elements with the current tag
        elements = doc.xpath('//%s' % tag)
        # Loop through each element and extract its text content
        for element in elements:
            extracted_text.append(element.text_content())
    # Join the extracted text into a single string and return it
    raw_text = ' '.join(extracted_text)
    raw_text = crt.normalize(raw_text)
    raw_text = crt.remove_stopwords(raw_text)
    crt.lemmatize(raw_text)
```
