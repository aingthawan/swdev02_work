```python
import pycountry
# print all contries from pycountry in lower case
countries = [country.name.lower() for country in pycountry.countries]
```

```python
# Import the required library
from geopy.geocoders import Nominatim
import pandas as pd

# Initialize Nominatim API
geolocator = Nominatim(user_agent="MyApp")
# dreate dataframe country and lat long
country_lat_long = pd.DataFrame(columns=['country', 'lat', 'long'])
for i in countries:
    try:
        location = geolocator.geocode(i)
        print(i)
        lat = location.latitude
        long = location.longitude
        print(lat, long)
        # concat to dataframe
        country_lat_long_temp = pd.DataFrame([[i, lat, long]], columns=['country', 'lat', 'long'])
        country_lat_long = pd.concat([country_lat_long, country_lat_long_temp], ignore_index=True)
    except:
        pass
```

```python
# export to csv
with open('country_lat_long.csv', 'w') as f:
    country_lat_long.to_csv(f, index=False)
```

```python
# get max and min values of the frequency
max_freq = max(country_dict.values())
min_freq = min(country_dict.values())
print(max_freq, min_freq)
```

```python
# for all countries in the dictionary, add a marker to the map
for country, freq in country_dict.items():
    # get the lat and long of the country
    lat = country_coordinate[country_coordinate['country'] == country]['lat'].values[0]
    long = country_coordinate[country_coordinate['country'] == country]['long'].values[0]
    folium.Marker([lat, long], popup=country.upper(), icon=folium.Icon(color="orange", icon='info-sign')).add_to(m)
```

```python
import folium
import branca.colormap

country_dict = {'india': 1, 'iceland': 29, 'ukraine': 173, 'spain': 50, 'turkey': 15, 'estonia': 100}

linear = folium.LinearColormap(
    ['yellow', 'orange', 'red'],
    vmin=min(country_dict.values()), vmax=max(country_dict.values()),
)

m = folium.Map(location=[0, 0], tiles='cartodbdark_matter', zoom_start=2)
# for all countries in the dictionary, add a marker to the map
for country, freq in country_dict.items():
    # get the lat and long of the country
    print(country, freq)
    lat = country_coordinate[country_coordinate['country'] == country]['lat'].values[0]
    long = country_coordinate[country_coordinate['country'] == country]['long'].values[0]
    folium.Marker([lat, long], popup=country.upper(), icon=folium.Icon(color=linear(freq), icon='info-sign')).add_to(m)
    
m
```
