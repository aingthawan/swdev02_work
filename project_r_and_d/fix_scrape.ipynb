```python
import re
from bs4 import BeautifulSoup
import requests

# url = "https://en.wikipedia.org/wiki/Main_Page"
url = "https://pantip.com/topic/41867493"

# Get the raw HTML data
response = requests.get(url)
html = response.content

# Create a BeautifulSoup object
soup = BeautifulSoup(html, "html.parser")

# Extract all the text from the HTML document
text = soup.get_text()
print(text)
```

```python
# text = re.sub(r"[^a-zA-Z0-9\s]", "", text)
text = re.sub(r"[^a-zA-Z0-9ก-๙เแไใๆ\s]+", "", text)
print(text)
```

```python
import requests

def checkAccessibility(url):
    """Check Whether URL is still accessible"""
    try:
        response = requests.get(url)
        print(response.status_code)
        print(response)
        response.raise_for_status()
        return True
    except (requests.exceptions.HTTPError, requests.exceptions.RequestException) as err:
        return False
```
